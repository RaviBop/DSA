```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
kmeans = KMeans(n_clusters=7)
```

```python
pop_df = pd.read_csv("./Pop_Data_numeric.csv")
pop_df.shape
```

```python
pop_df
```

```python
cluster_df = pop_df[['Age', 'Take_Home_Pay']]#, 'Children_cat']]
cluster_df.shape
```

```python
cluster_df = cluster_df.dropna()
cluster_df.shape
```

```python
kmeans.fit(cluster_df)
```

```python
cluster_df['cluster_id'] = kmeans.labels_
```

```python
plt.scatter(cluster_df['Age'], cluster_df['Take_Home_Pay'], c=cluster_df['cluster_id'], s=5, cmap='viridis')
plt.xlabel('age')
plt.ylabel('take home pay')
plt.show()
```

```python
kmeans = KMeans(n_clusters=30)
cluster_df = pop_df[['Age', 'Take_Home_Pay', 'Children_cat']]
cluster_df = cluster_df.dropna()
kmeans.fit(cluster_df)
cluster_df['cluster_id'] = kmeans.labels_
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(cluster_df['Age'], cluster_df['Children_cat'], cluster_df['Take_Home_Pay'], c=cluster_df['cluster_id'], s=5, cmap='viridis')
plt.xlabel('age')
plt.ylabel('children')
ax.set_zlabel('take home pay')
plt.show()
```

```python
kmeans = KMeans(n_clusters=20)
cluster_df = pop_df[['Age', 'Take_Home_Pay', 'Government_Office_Region_cat']]
cluster_df = cluster_df.dropna()
kmeans.fit(cluster_df)
cluster_df['cluster_id'] = kmeans.labels_
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(cluster_df['Age'], cluster_df['Government_Office_Region_cat'], cluster_df['Take_Home_Pay'], c=cluster_df['cluster_id'], s=5, cmap='viridis')
plt.xlabel('age')
plt.ylabel('Government_Office_Region_cat')
ax.set_zlabel('take home pay')
plt.show()
```
